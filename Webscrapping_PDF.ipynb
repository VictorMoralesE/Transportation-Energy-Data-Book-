```python
!pip install PyMuPDF
import requests
from bs4 import BeautifulSoup
import os
import re
import fitz  # PyMuPDF

# URL de la página de archivo de Transportation Energy Data Book
url = 'https://tedb.ornl.gov/archive/'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar todos los enlaces que contienen el texto "Transportation Energy Data Book"
links = soup.find_all('a', string=lambda text: "Transportation Energy Data Book" in text if text else False)

# Regex para identificar ediciones entre 28 a 40
edition_regex = re.compile(r'(\d{2,})', re.I)

# Función para extraer datos de la tabla del PDF
def extract_data_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    table_data = []
    for page in doc:
        text = page.get_text()
        if "Table 2.5" in text and "Domestic Consumption of Transportation Energy by Mode and Fuel Type" in text:
            # Here you should extract data from the table, but for this example, let's just print the page text
            table_data.append(text)
    return table_data

for link in links:
    # Extraer la URL del archivo PDF
    pdf_url = link.get('href')
    # Extraer el nombre del archivo PDF
    pdf_name = pdf_url.split('/')[-1]

    # Obtener el número de edición del PDF
    edition_match = edition_regex.search(pdf_name)
    if edition_match:
        edition_number = int(edition_match.group(1))
        # Verificar si la edición está entre 28 y 40
        if 28 <= edition_number <= 40:
            # Realizar la solicitud para descargar el archivo PDF
            pdf_response = requests.get(pdf_url)
            # Guardar el archivo PDF en el directorio actual
            with open(pdf_name, 'wb') as file:
                file.write(pdf_response.content)
            print(f'Descargado: {pdf_name}')

            # Extraer datos de la tabla del PDF
            table_data = extract_data_from_pdf(pdf_name)
            for page_text in table_data:
                print(page_text)

print('Descarga y extracción completadas.')
```
